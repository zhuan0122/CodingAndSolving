给出两个字符串 s 和 t，要求在 s 中找出最短的包含 t 中所有字符的连续子串。

数据范围：0>∣S∣,∣T∣≤100000 > |S|,|T| \le10000 0>∣S∣,∣T∣≤10000，保证s和t字符串中仅包含大小写英文字母
要求：进阶：空间复杂度 O(n)O(n)O(n) ， 时间复杂度 O(n)O(n)O(n)
例如：

S="XDOYEZODEYXNZ"S ="XDOYEZODEYXNZ"S="XDOYEZODEYXNZ"
T="XYZ"T ="XYZ"T="XYZ"
找出的最短子串为"YXNZ""YXNZ""YXNZ".

注意：
如果 s 中没有包含 t 中所有字符的子串，返回空字符串 “”；
满足条件的子串可能有很多，但是题目保证满足条件的最短的子串唯一。


```python
### my solution. 
** out of time limit 超时！！！！ 
#
# 代码中的类名、方法名、参数名已经指定，请勿修改，直接返回方法规定的值即可
#
# 
# @param S string字符串 
# @param T string字符串 
# @return string字符串
import math
class Solution:
    def minWindow(self , S: str, T: str) -> str:
        # write code here
        # 什么是滑动串口 肯定是有左右边界的 window 和最长回文字串  怎样
        # 我的想法 是 不在乎 顺序 只要出现过所 滑动串口大小 可以至少是 T 的长度 因为至少要包含T 中所有元素
        # 滑动窗口的window 长度 也是 递增的 最小从T 长度开始 
        if len(T)==0:
            return ''
        if S==T:
            return T
        # least length
        window_s= len(T)
        window_l=len(S)
        [start, end] = [-math.inf, math.inf]
        for w in range(window_s,window_l):
            # for each window interval, we match them with T
            #print('w',w)
            for i in range(len(S)):
                if i+w>len(S):
                    break
                #print('check',[i,w], S[i:i+w])
                if self.checkWindow(i, w, S, T):
                    #print('find it')
                    if w<(end-start):
                        [start, end] = [i,i+w] 
        return S[start:end] if [start, end] != [-math.inf, math.inf] else ''
            
                
                
    def checkWindow(self,begin, window,S,T):
        cache_w = {}
        cache_t= {}
        #print('here', [begin, window])
        for i in range(begin,begin+window):
            if S[i] not in cache_w:
                cache_w[S[i]]= 1
            else:
                cache_w[S[i]]+=1
        #print('cache_w', cache_w)
        for i in range(len(T)):
            if T[i] not in cache_t:
                cache_t[T[i]]=1
            else:
                cache_t[T[i]]+=1
        #print('cache_t',cache_t)
                
        for i in cache_t:
            #print('i',i)
            if i not in cache_w:
                return False 
            else:
                if cache_t[i]!=cache_w[i]:
                    return False 
        return True
```

### 2. Sliding windows:
** 1. i, j 左右边界都包括的 窗口。 外循环对J 外括， 找到满足条件的一个 子结果 第二步 在子结果的基础上 对I 进行 收缩， 找到最终子结果
** 注意边界条件 和什么时候收缩以及 更新 NEED 列表中的元素对应的个数以及总的 NEED 需要的个数。所以一开始我们将 所需要元素存入 哈希表 并且
记录需要的个数，。 以及总体需要的个数 和开始I 为位置 0 




```python
#
# 代码中的类名、方法名、参数名已经指定，请勿修改，直接返回方法规定的值即可
#
# 
# @param S string字符串 
# @param T string字符串 
# @return string字符串
import math
class Solution:
    def minWindow(self , S: str, T: str) -> str:
        # write code here
        # 滑动窗口 我们用到 左右两个区间 左右闭合。 所以窗口内长度为 j-i+1. 
        # i , j 从0 开始 先扩张 J 直到 我们判断 目前区间包含了所有我们需要的 元素。 怎么判断 我们
        # 借助哈希表 先把 需要的所有元素 存入哈希表中， 当我们在 扩张 窗口的时候 我们同时对哈希表进行动态维护。
        # 每当我们碰到需要元素就 将哈希表中对应元素 减去1 并且 CNT 减去1. 当CNT 到0 时候我们可以知道
        #目前窗口不用继续扩张 因为已经包含所有元素。 
        # 接下来要做的就是对窗口进行收缩。因为我们找到这个窗口中 最小的那个 即最小的包含所有元素的窗口。并且记录当前
        #窗口的起始位置
        ## 滑动窗口的时间复杂度 考虑的时候 主要是 窗口的扩张 是O*(N) 但是当前窗口确定后， 窗口的 收缩 也是 一次
        # 循环扫描 我觉得最坏情况下是O*(N N)
        if len(T)==0:
            return ''
        
        res=[0, math.inf]
        
        # 1. store all needed elements in hash table 
        need={}
        # 记录我们需要的字符的长度因为要考虑重复元素 但是不用考虑位置
        needCnt=len(T)
        for i in T:
            if i not in need:
                need[i]=1 
            else:
                need[i]+=1
        # 初始窗口的左窗口界值
        i=0
        
        for j in range(len(S)):
            # 2 .expand window with j moving to right
            # J 是往右扩张的 直到 包含了所有 NEED 中的元素 且要求个数相同。 
            # 所以 当J 在右移动过程中 如果我们遇到了 NEED 中元素 我们考虑的是如果这个元素当前个数大于1 那么就对总的
            # 需要的个数NEEDCNT 减去1.
            # 但是因为 我们也计算重复出现的个数 所以 S》 XXXYZ， T》 XYZ  
            #我们也要及时减去1 只要它出现在NEED 中。 但是 NEEDCNT
            # 只在 当前这个元素 的 个数是大于0 的时候 就是 第一个X 出现的时候 我们知道了
            # X 出现了 NEEDCNT 减去1. 对应该元素个数也变成0. 第二个X 出现时 我们不会再将NEEDCNT 减去1 
            ## 因为 我们只需要1 个X。 我们还需要 YZ 还是需要两个
            if S[j] in need:
                if need[S[j]]>0:
                    # for example we only need 1 X in T, but we have 3 X in S before we have all needed 
                    #elments. we do not make needCnt minus 1 unless its counted number is bigger than 1 
                    needCnt-=1 
                need[S[j]]-=1 
                
            # 3.当 需要的元素全部出现完毕 NEEDCNT 是0 那么 我们就需要 对当前窗口 进行收缩。 因为我们可能不需要
            # 当前所有元素。 所以我们需要收缩窗口 对I 进行右移动
            if needCnt==0:
                # move i to right to minimize the window 
                # 我们停止右移 I 当我们 发现了 第一个 元素 这个元素在NEED 哈希表中
                while True:
                    ## 如果 当前 I 在哈希表中 那么我们就进一步判断他是不是 对应个数为0 如果是的那就是表明我们
                    # 找到了 当前的窗口 且没有多余的超过我们需要的的个数的重复元素。 这里比较重要 因为 如果当前元素
                    #是在NEED 列表中 但是 它个数不是0 那么说明 当前 是多余的 因为即使我们需要2个X 但是如果当前X 
                    ##位置会包括3 个X 那么还不是最短的满足条件的窗口我们可以进一步收缩
                    if S[i] in need:
                        
                        if need[S[i]]==0:
                            # break while loop
                            break
                        # this means we have more S[I] then we need 
                        # 元素在NEED 中 但是个数多余 我们需要进一步收缩 但是把这个元素移除 NEED 中 就是 
                        # 这个元素对应的个数加上1 
                        need[S[i]]=need[S[i]]+1 
                    # 当前  S[i] 在不在 列表中 只要没有满足 BREAK 条件 我们都要把他加 收缩一步
                    i+=1
                        
                if j-i < res[1]-res[0]: 
                    # store current applicable window 
                    res=(i,j)
                    
                # 4.update need hash table for looking for next window 
                # 在我们确定了 当前符合条件的窗口之后 我们需要开始新的滑动窗口 就是 J 会继续右移动 但是
                # I 的起始位置要加1 因为 我们会把 当前I 位置所指元素重新加入 NEED 列表中
                need[S[i]]+=1
                needCnt+=1 
                i+=1 
        return '' if res[1]>len(S) else S[res[0]:res[1]+1]
```
